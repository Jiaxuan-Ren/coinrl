```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/Gemini_BTCUSD_1h.csv')
df = df.sort_values('Date')
df = df.drop(columns=["unix", "Volume_BTC", "Symbol"]).reset_index(drop=True)
```

```python
df = df.drop(list(range(5))).reset_index(drop=True)
```

```python
pnl_list = [1, 5, 10, 100]

for pnl_period in pnl_list:
    df_pnl_key = "pnl_{}".format(pnl_period)
    df[df_pnl_key] = df.Close.shift(-pnl_period) - df.Close
df
```

```python
df = df.fillna(0)
```

```python
df.to_csv("data/BTCUSD_1h_pnl.csv")
```
